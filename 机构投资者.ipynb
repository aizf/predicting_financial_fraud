```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_excel('./data/2.xlsx',sheet_name='Sheet3',
keep_default_na=False,
    dtype={
        'COMPANY': str,
        'DATE': str,
        'RATIO': float,
    })

df1=df1.dropna()
df1.head()
# df1["AV_RATIO"]=None
```

```python
print(df1.isnull().sum())
print(df1.shape)
print(df1.info())
# df1.iloc[31520:]
# df1[df1["GAIN"] > 0.5]
```

```python
COMS=[]
for index, row in df1.iterrows():
        COMS.append(row["COMPANY"])
COMS=list(set(COMS))
dfs=[]
for COMPANY in COMS:
    dfs.append(df1[df1["COMPANY"]==COMPANY])
```

```python
AV_RATIO={}
for df in dfs:
    COMPANY=df["COMPANY"].iloc[0]
    YEARS=[]
    for index, row in df.iterrows():
        # print(row["DATE"].split("-"))
        YEARS.append(row["DATE"].split("-")[0])
    YEARS=list(set(YEARS))1
    for YEAR in YEARS:
        df_YEAR=df[df["DATE"].str.contains(YEAR)]
        AV_RATIO[str(int(YEAR)*10000000+int(COMPANY))]=df_YEAR["RATIO"].mean()


import json
with open("./data/AV_RATIO.json", 'w') as f:
    json.dump(AV_RATIO, f)
```
